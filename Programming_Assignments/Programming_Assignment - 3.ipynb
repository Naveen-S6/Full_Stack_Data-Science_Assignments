### 1. Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
def CheckNumber(n):
    if n>0:
        print("The number is a postive number")
    elif n<0:
        print("The number is a negative number")
    else:
        print("It is zero")
```

```python
CheckNumber(10)
```

```python
CheckNumber(-20)
```

```python
CheckNumber(0)
```

### 2. Write a Python Program to Check if a Number is Odd or Even?

```python
def OddOrEven(num):
    if num % 2 == 0:
        print("The number is even")
    else:
        print("The number is odd")
```

### 3. Write a Python Program to Check Leap Year?

To check if a year is a leap year, divide the year by 4. If it is fully divisible by 4, it is a leap year.

Century years like 300, 700, 1900, 2000 need to be divided by 400

```python
def CheckLeap(year):
    if year % 4 == 0:
        str_year = str(year)
        if str_year[-3:] =='000'or str_year[-2:] =='00':
            if year % 400 == 0:
                print("It is a leap year")
            else:
                print("It is not a leap year")
        else:
            print("It is a leap year")
    else:
        print("It is not a leap year")
                
```

```python
CheckLeap(2000)
```

```python
CheckLeap(700)
```

```python
CheckLeap(1300)
```

```python
CheckLeap(2022)
```
